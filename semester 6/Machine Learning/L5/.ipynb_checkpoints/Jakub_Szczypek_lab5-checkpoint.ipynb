**Ex. 1 - kNN - own implementation**

```python
def euclidean_distance(x1, x2):
    distance = np.sqrt(np.sum((x1-x2)**2)) 
    return distance 

class KNN:
    def __init__(self, k=3):
        self.k = k
        
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return predictions
        
    def _predict(self, x):
        #compute the distance
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        
        #get the closest k
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]
        
        #mojority voye
        most_common =np.argmax(np.bincount(np.array(k_nearest_labels)))
        return most_common
```

```python
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 0, test_size=0.3)

clf = KNN(k=5)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
acc = np.sum(predictions == y_test) / len(y_test)
print(acc)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 0, test_size=0.3)

k_tab = range(1, len(X_train))
train_errors = []
test_errors = []
for k in k_tab:
    knn = KNN(k)
    knn.fit(X_train, y_train)
    y_test_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_test_pred)
    test_errors.append(accuracy)
    y_train_pred = knn.predict(X_train)
    accuracy = accuracy_score(y_train, y_train_pred)
    train_errors.append(accuracy)
    
plt.scatter(k_tab, test_errors, label = "Test error")
plt.scatter(k_tab, train_errors, label = "Train error")
plt.title('Accuracy dependence on k')
plt.ylabel('accuracy')
plt.legend()
plt.xlabel('k')
plt.show()

print(f"Wartość k dla największej dokładności: {k_tab[np.argmax(test_errors)]} i wynosi ona {max(test_errors)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn import tree
from sklearn.model_selection import train_test_split
from scipy.io import loadmat
```

```python
dane = loadmat('mnist.mat')
```

```python
X = dane["trainX"]
y = dane["trainY"]

# Standaryzacja
for i in range(X.shape[0]):
    X[i,:] = X[i,:]/np.std(X[i,:])
```

```python
def plot_mnist(images, titles, h, w, n_row=3, n_col=4):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.05)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        # tu usunalem transpozycje przy reshape()
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
```

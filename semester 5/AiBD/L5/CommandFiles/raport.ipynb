## Modyfikacja danych

Usunąłem wszystkie kraje, w których spożycie alkoholu jest równe zero:

```python
drinks_without_zero_consumption = drinks.loc[drinks['total_litres_of_pure_alcohol'] != 0]

#Zapis zmodyfikowanych danych do pliku dirnks_without_zero_consumption
filepath = Path('../AnalysisData/drinks_without_zero_consumption.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)
drinks_without_zero_consumption.to_csv(filepath) 

# Wczytanie nowych danych
drinks_new = pd.read_csv('../AnalysisData/drinks_without_zero_consumption.csv')

print("\nUsunięto ze zbioru {} krajów.".format(drinks.index[-1] - drinks_new.index[-1]))
```

## Data appendix

```python
drinks_new.drop(columns=['Unnamed: 0'], inplace=True)

print("Podstawowe informacje o bazie:")
drinks_new.info()
```

```python
print("Indeksy w całym zbiorze danych:")
print(drinks_new.index[0], " - ",  drinks_new.index[-1])

print("\n")
print("Spożycie piwa:")
print(drinks_new["beer_servings"].min(), " - ",  drinks_new["beer_servings"].max())

print("\n")
print("Spożycie spirytusu:")
print(drinks_new["spirit_servings"].min(), " - ",  drinks_new["spirit_servings"].max())

print("\n")
print("Spożycie wina:")
print(drinks_new["wine_servings"].min(), " - ",  drinks_new["wine_servings"].max())

print("\n")
print("Całkowite spożycie czystego alkoholu:")
print(drinks_new["total_litres_of_pure_alcohol"].min(), " - ",  drinks_new["total_litres_of_pure_alcohol"].max())
```

## Wykresy

```python
beer_mean = drinks_new["beer_servings"].mean()
spirit_mean = drinks_new["spirit_servings"].mean()
wine_mean = drinks_new["wine_servings"].mean()
alc_mean = drinks_new["total_litres_of_pure_alcohol"].mean()

means = [beer_mean, spirit_mean, wine_mean, alc_mean]
names = ["beer_servings", "spirit_servings", "wine_servings", "total_litres_of_pure_alcohol"]

plt.bar(names, means)
plt.xlabel('Rodzaj alkoholu')
plt.ylabel('Ilość spożycia')
plt.grid()
plt.title('Średnie spożycie różnych alkoholi')
plt.show()
```

```python
high_beer_consumption = drinks_new[drinks_new['beer_servings'] > 290]
high_spirit_consumption = drinks_new[drinks_new['spirit_servings'] > 230]
high_wine_consumption = drinks_new[drinks_new['wine_servings'] > 200]
high_alcohol_consumption = drinks_new[drinks_new['total_litres_of_pure_alcohol'] > 11]

ax1 = high_beer_consumption.plot.bar(x='country', y='beer_servings', color = 'Violet')
ax1.set_title('Kraje z największym spożyciem piwa')
ax1.grid()
ax1.set_ylim(290, 380)
ax1.set_ylabel('Ilość puszek piwa')
ax1.set_xlabel('Nazwy krajów')

ax2 = high_spirit_consumption.plot.bar(x='country', y='spirit_servings', color = 'blue')
ax2.set_title('Kraje z największym spożyciem spirytusu')
ax2.grid()
ax2.set_ylim(230, 450)
ax2.set_ylabel('Ilość kieliszków spirytusu')
ax2.set_xlabel('Nazwy krajów')

ax3 = high_wine_consumption.plot.bar(x='country', y='wine_servings', color = 'green')
ax3.set_title('Kraje z największym spożyciem wina')
ax3.grid()
ax3.set_ylim(200, 380)
ax3.set_ylabel('Ilość kieliszków wina')
ax3.set_xlabel('Nazwy krajów')

ax4 = high_alcohol_consumption.plot.bar(x='country', y='total_litres_of_pure_alcohol', color = 'black')
ax4.set_title('Kraje z największym ogólnym spożyciem alkoholu')
ax4.grid()
ax4.set_ylim(11, 15)
ax4.set_ylabel('Ilość spożytego alkoholu w litrach')
ax4.set_xlabel('Nazwy krajów')
```

**Zadanie 3**

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)

y1 = x1**2 + 5
y2 = x2**2 + 5
y3 = x3**2 + 5

plt.plot(x1, y1)
plt.title("Wykres f(x) = x^2 + 5 w przedziale <-1,1>")
plt.grid()
plt.xlabel("O x")
plt.ylabel("O y")
plt.legend("f1")
plt.show()

plt.plot(x2, y2)
plt.title("Wykres f(x) = x^2 + 5 w przedziale <-6,6>")
plt.grid()
plt.xlabel("O x")
plt.ylabel("O y")
plt.legend("f2")
plt.show()

plt.plot(x3, y3)
plt.title("Wykres f(x) = x^2 + 5 w przedziale <0,5>")
plt.grid()
plt.xlabel("O x")
plt.ylabel("O y")
plt.legend("f3")
plt.show()
```

**W trakcie zadania 3 stworzyem 3 wykresy funkcji f(x) = x^2 + 5 dla rnych parametrw x**

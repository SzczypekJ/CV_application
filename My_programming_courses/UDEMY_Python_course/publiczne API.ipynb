## Otwieranie stron

```python
import requests
import json
import pprint
import webbrowser


# Nie było żadnego pytania ze score większym od 12 więc ustawiłem sobie 10 zamiast 15
params = {
    "site" : "stackoverflow",
    "sort" : "votes",
    "order" : "desc",
    "fromdate" : "2023-10-24",
    "tagged" : "python",
    "min" : 10
}

r = requests.get("https://api.stackexchange.com/2.3/questions/", params)

try:
    questions = r.json()
except json.decoder.JSONDecodeError:
    print("Nie poprawny format")
else:
    for question in questions["items"]:
        webbrowser.open_new_tab(question["link"])
```

## Ulepszenie powyższego kodu - automatyczna generacja czasu

```python
from datetime import datetime, timedelta

time_before = timedelta(days=7)
searchdate = datetime.today() - time_before

# Nie było żadnego pytania ze score większym od 12 więc ustawiłem sobie 10 zamiast 15
params = {
    "site" : "stackoverflow",
    "sort" : "votes",
    "order" : "desc",
    "fromdate" : int(searchdate.timestamp()),
    "tagged" : "python",
    "min" : 10
}

r = requests.get("https://api.stackexchange.com/2.3/questions/", params)

try:
    questions = r.json()
except json.decoder.JSONDecodeError:
    print("Nie poprawny format")
else:
    for question in questions["items"]:
        webbrowser.open_new_tab(question["link"])
```

## Funkcja pobierająca z API randomowe zdjęcia psów i wyświetlająca je w przeglądarce

```python
import json
import requests
from pprint import pprint
import webbrowser


userNumberChoice = int(input("Give me the number of dog images u want to see: "))

def number_of_dog_images(amount):


    r = requests.get("https://dog.ceo/api/breeds/image/random/" +  str(amount))


    try:
        content = r.json()
    except json.decoder.JSONDecodeError:
        print("Nie poprawny format")
    else:
        for image in content["message"]:
            webbrowser.open_new_tab(image)

number_of_dog_images(userNumberChoice)
```

## Headers, POST, DELETE

**Pobieranie loginu i hasła, sprawdzamy czy login i hasło są poprawne, logowanie zaszło pomyślnie, pobieramy z bazy danych userID i name - nazwe użytkownika**

```python
import requests
import json
from pprint import pprint
import credentials

def get_json_content_from_response(response):
    try:
        content = response.json()
    except json.decoder.JSONDecodeError:
        print("Invalid format", response.text)
    else:
        return content


def get_favourites_cats(userID):
    params = {
        "sub_id" : userID
    }

    r = requests.get("https://api.thecatapi.com/v1/favourites", params, headers=credentials.headres)

    return get_json_content_from_response(r)

def get_random_cat():
    r = requests.get("https://api.thecatapi.com/v1/images/search", headers=credentials.headres)

    return get_json_content_from_response(r)[0]

def add_favourite_cat(catId, userId):
    cat_data = {
        "image_id": catId,
        "sub_id": userId
    }
    r = requests.post("https://api.thecatapi.com/v1/favourites", json=cat_data, headers=credentials.headres)

    return get_json_content_from_response(r)


def delete_cat_from_favourites(userId, favourite_cat_id):
    r = requests.delete("https://api.thecatapi.com/v1/favourites/"+favourite_cat_id, headers=credentials.headres)

    return get_json_content_from_response(r)

print("Hello, log in - give me your login and password")

userId = "agh2m"
name = "Jakub"

print("Hello " + name)
favourite_cats = get_favourites_cats(userId)
print("Your favourite cats are:", favourite_cats)
random_cat = get_random_cat()
print("You have drawn the following kitten: ", random_cat["url"])

add_to_favourites = input("Do you want to add it to your favourites? T/F")

newly_added_cat_info = {}
if add_to_favourites.upper() == "T":
    resoult_from_adding_favourite_cat = add_favourite_cat(random_cat["id"], userId)
    newly_added_cat_info = {resoult_from_adding_favourite_cat["id"] : random_cat["url"]}
else:
    print("You didnt add a cat to your favourites, cat will be upset :(")

favourite_cat_by_id = {
    favourite_cat["id"] : favourite_cat["image"]["url"]
    for favourite_cat in favourite_cats
}

favourite_cat_by_id.update(newly_added_cat_info)

print(favourite_cat_by_id)
favourite_cat_id = input("Do you want to delete cat from favourites? Give me the id")
print(delete_cat_from_favourites(userId, favourite_cat_id))
```

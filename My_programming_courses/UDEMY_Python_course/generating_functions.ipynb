## Funckje generujące

```python
# Wyrażenie generujące

evenNumberGenerators = (element
    for element in range (400)
    if element % 2 == 0)

print(evenNumberGenerators)
```

```python
# Funkcje generujące

def generate_even_numbers():
    for element in range(400):
        if element % 2 == 0:
            yield element

a = generate_even_numbers()
# next(a) - służy do wywoływania kolejnych elementów generatora

def generate_10_numbers():
    x = 0
    while x < 10:
        yield x
        x += 1

print(list(generate_10_numbers()))
print(list(generate_10_numbers()))
print(list(generate_10_numbers()))
print(list(generate_10_numbers()))

generate_10_numbers_expression = (x
                                  for x in range (10))

print(list(generate_10_numbers_expression))
print(list(generate_10_numbers_expression))
```

## Funkcja generująca kwadraty liczb w nieskończoność. Korzystając z funkcji generującej wygeneruj 20 elementów, po czym wróć do momentu skończenia i wygeneruj następnie 30 kolejnych liczb. Zapisz wygenerowane elementy w tej samej liście.

```python
def number_multilied_by_itself_generator():
    number = 0
    while True:
        print("Start number", number)
        number = number + 1
        yield number * number
        print("Po yield number", number)

generateNumbers = []
numberGenerator = number_multilied_by_itself_generator()
for _ in range(20):
    generateNumbers.append(next(numberGenerator))

print(generateNumbers)


for _ in range(30):
    generateNumbers.append(next(numberGenerator))

print(generateNumbers)
```

## Metoda send - wysyłanie wiadomości do generatora

```python
def number_multilied_by_itself_generator():
    number = 0
    while True:
        print("Start number", number)
        # number = number + 1
        number = yield number * number
        print("Po yield number", number)


generateNumbers = []
numberGenerator = number_multilied_by_itself_generator()

numberGenerator.send(None)
# next(numberGenerator)
for i in range(20):
    generateNumbers.append(numberGenerator.send(i))

print(generateNumbers)


for i in range(30, 50):
    generateNumbers.append(numberGenerator.send(i))

print(generateNumbers)
```
